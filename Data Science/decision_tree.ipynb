```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/DFE5/main/breastcancer_data.csv')
data.head()
```

```python
columns = list(data.columns[2:])
```

```python
for col1 in columns:
    for col2 in columns:
        print(f"Correlation between {col1} and {col2} is {data[col1].corr(data[col2])}")
```
